```python
from torch.utils.data import Dataset

class MultimodalDataset(Dataset):
    def __init__(self, csv_data, image_dir, tokenizer, max_len, transform=None):
        self.csv_data = csv_data
        self.image_dir = image_dir
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.transform = transform

    def __len__(self):
        return len(self.csv_data)

    def __getitem__(self, idx):
        row = self.csv_data.iloc[idx]
        img_path = f"{self.image_dir}/{row['image_id']}.jpg"  # Adjust extension if needed
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        text = row['transcriptions']
        text_inputs = self.tokenizer(
            text,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        label = torch.tensor(row['labels'], dtype=torch.float)
        return image, text_inputs['input_ids'].squeeze(0), text_inputs['attention_mask'].squeeze(0), label
```

```python
import torch
from PIL import Image
import pandas as pd
from torch.utils.data import Dataset
from transformers import BertTokenizer

# Initialize the tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
max_len = 128
```

```python
from torchvision import transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize image to 224x224
    transforms.ToTensor(),         # Convert image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize for pre-trained models
])
```

```python
# Example usage
train_dataset = MultimodalDataset(train_csv, '/content/Train/images', tokenizer, max_len, transform)
dev_dataset = MultimodalDataset(dev_csv, '/content/Dev/images', tokenizer, max_len, transform)
test_dataset = MultimodalDataset(test_csv, '/content/Test/images', tokenizer, max_len, transform)
```

```python
import pandas as pd

# Load the datasets (adjust the file path as needed)
train_csv = pd.read_csv('/content/Train/train.csv')
dev_csv = pd.read_csv('/content/Dev/dev.csv')
test_csv = pd.read_csv('/content/Test/test.csv')
test_labels_csv=pd.read_csv('/content/Test_labels/test_with_labels.csv')

# Check dataset columns (ensure there's a column 'transcriptions' and 'labels')
train_csv.head()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load Malayalam BERT model and tokenizer
#model_name = "ai4bharat/indic-bert"  # Replace with your desired BERT model
model_name = "l3cube-pune/malayalam-bert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)  # Binary classification

# Tokenize the transcriptions
def tokenize_data(data, tokenizer, max_length=128):
    return tokenizer(
        list(data['transcriptions'].fillna("")),
        padding=True,
        truncation=True,
        max_length=max_length,
        return_tensors="pt"
    )


train_encodings = tokenize_data(train_csv, tokenizer)
dev_encodings = tokenize_data(dev_csv, tokenizer)
test_encodings = tokenize_data(test_csv, tokenizer)

import numpy as np
from sklearn.utils.class_weight import compute_class_weight

# Convert the list of classes to a numpy array
classes = np.array([0, 1])

# Compute class weights (if there's class imbalance)
class_weights = compute_class_weight('balanced', classes=classes, y=train_csv['labels'])
class_weights = torch.tensor(class_weights, dtype=torch.float)

# Print class weights for verification
print(f"Class Weights: {class_weights}")
import torch

# Convert labels to tensor
train_labels = torch.tensor(train_csv['labels'].values)
dev_labels = torch.tensor(dev_csv['labels'].values)
from torch.utils.data import Dataset

class MalayalamBERTDataset(Dataset):
    def __init__(self, encodings, labels=None):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.encodings["input_ids"])

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        if self.labels is not None:
            item["labels"] = self.labels[idx]
        return item

train_dataset = MalayalamBERTDataset(train_encodings, train_labels)
dev_dataset = MalayalamBERTDataset(dev_encodings, dev_labels)
test_dataset = MalayalamBERTDataset(test_encodings)  # No labels for test
import os
from transformers import Trainer, TrainingArguments

# Disable W&B if you don't want to use it
os.environ["WANDB_DISABLED"] = "true"

training_args = TrainingArguments(
    output_dir="./results",
    run_name="malayalam-bert-training",  # Optional: Set a unique name for W&B runs
    eval_strategy="epoch",  # Updated to avoid deprecation warning
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=10,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=dev_dataset,
)




```

```python
test_results = trainer.predict(test_dataset)
predictions = torch.argmax(torch.tensor(test_results.predictions), dim=1)

# Save predictions
test_csv['predicted_labels'] = predictions.numpy()
test_csv.to_csv('/content/Test/test_predictions.csv', index=False)


# Reload and view the saved file
predicted_file = pd.read_csv('/content/Test/test_predictions.csv')
predicted_file.head() # Display the first few rows
import pandas as pd

# Load the predicted file and true labels file
predicted_file = pd.read_csv('/content/Test/test_predictions.csv')
true_labels_file = pd.read_csv('/content/Test_labels/test_with_labels.csv')

# Display the first few rows of both files to verify
print("Predicted File:")
predicted_file.head()
print("True Labels File:")
true_labels_file.head()
# Merge the true labels and predictions based on 'image_id'
merged_df = pd.merge(true_labels_file, predicted_file, on='image_id', suffixes=('_true', '_pred'))

# Display the merged dataset
merged_df.head()
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Extract true labels and predicted labels
true_labels = merged_df['labels']  # Column with true labels
predicted_labels = merged_df['predicted_labels']  # Column with predicted labels

# Calculate metrics
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels, average='binary')  # Use 'macro' or 'weighted' for multi-class
recall = recall_score(true_labels, predicted_labels, average='binary')
f1 = f1_score(true_labels, predicted_labels, average='binary')

# Print metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
merged_df.to_csv('/content/Test/comparison_results.csv', index=False)
print("Comparison results saved as '/content/Test/comparison_results.csv'")
```
